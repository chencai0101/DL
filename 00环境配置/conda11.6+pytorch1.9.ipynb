## conda11.6+pytorch+python 3.8

版本契合很重要！

具体安装方式参考：https://blog.csdn.net/weixin_42421914/article/details/132870442

简单说就是：
1. 查看conda版本`nvcc -V`
2. 安装GPU版本的PyTorch：打开anaconda Prompt，切换至pytorch环境，输入以下代码。
```
conda install pytorch==1.13.1 torchvision==0.14.1 torchaudio==0.13.1 pytorch-cuda=11.6 -c pytorch -c nvidia
```
安装完成后执行如下代码验证cuda是否可用

````python
import torch

torch.cuda.is_available()
````
