### 矢量化加速实验

#### 以下是一个记录运行时间的类。
1. 其中init函数表示类被实例化时会执行的操作，
2. 每个函数都传入 self 是因为这样才可以指向类实例。

```python
import math 
import time 
from mxnet import np 
from d2l import mxnet as d2

class Timer: #@save 
    """记录多次运行时间""" 
    def __init__(self): 
        self.times = [] 
        self.start() 
    
    """启动计时器""" 
    def start(self): 
        self.tik = time.time() 
    
    """停止计时器并将时间记录在列表中"""
    def stop(self):  
        self.times.append(time.time() - self.tik) 
        return self.times[-1] 
    
    """返回平均时间""" 
    def avg(self):
        return sum(self.times) / len(self.times) 

    """返回时间总和"""
    def sum(self): 
        return sum(self.times) 
    
    """返回累计时间"""
    def cumsum(self):  
        return np.array(self.times).cumsum().tolist()
```

#### 方案一：用for循环实现两个向量相加

```python
timer = Timer()
for i in range(100):
    c[i]=a[1]+b[i]

f'{timer.stop():.5f}'    # 格式化字符串的操作，表示将结果以保留小数点后五位的浮点数形式显示。
```
