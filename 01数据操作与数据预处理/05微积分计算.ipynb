## 微积分计算

### 导数与微分实验

#### 定义函数并计算其导数

$$f(x)=3x^2-4x$$
$$f^{\prime}(x)=\lim _{h \rightarrow 0} \frac{f(x+h)-f(x)}{h}$$

```python
def f(x):                        # 函数
    return 3 * x ** 2 - 4 * x

def numerical_lim(f, x, h):      # 极限
    return (f(x + h) - f(x)) / h

k= numerical_lim(f,2,0.1)        # 计算f'(x)在 x=2 这一点的导数，这里的 Δx 取的0.1
```

#### 绘制函数与导数图像

```python
import matplotlib.pyplot as plt
import numpy as np

# 准备数据
x = np.linspace(-5, 5, 100)
y1=f(x)
y2=k*(x-2)+f(2)

plt.plot(x, y1, label='second-degree function')          # 二次函数
plt.plot(x, y2, label='tangent line')                    # 切线

# 添加标题和坐标轴标签
plt.title('experiment on derivatives')
plt.xlabel('X')
plt.ylabel('Y')

# 设置图例
plt.legend()

# 显示图形
plt.show()
```

### 自动微分
> 自动微分的“自动”是相交于“导数与微分实验”中我们通过导数定义来模拟的微分而言的；`pytorch`提供了一个现成的微分函数；

```python
import torch

# 创建一个简单的线性模型参数
w = torch.randn(1, requires_grad=True)  # requires_grad=True，意味着每次更新梯度时，不会分配新内存。
b = torch.randn(1, requires_grad=True)

# 输入数据
x = torch.tensor([2])

# 前向传播计算输出
y = w * x + b

# 定义损失函数（均方误差）
loss = (y - 5.) ** 2

# 执行反向传播
loss.backward()

# 查看参数的梯度
print("w 的梯度:", w.grad)
print("b 的梯度:", b.grad)
```
