## 张量转化为其他Python对象
### 张量与Numpy张量相互转化

```python
import torch

X=torch.tensor([2,3,4])

A = X.numpy()         # pytorch张量转化为Numpy张量
B = torch.tensor(A)   # Numpy张量转化为pytorch张量

type(A)
type(B)
```

### 大小为1的张量转化为Python标量

```python
a = torch.tensor([3.5])
int(a)    # 转为整数，会自动取整
float(a)  # 转为浮点数
```
