### 正态分布

定义正态分布函数

```python
import torch
from torch.distributions import Normal

# 定义正态分布的参数
mu = 0.0  # 均值
sigma = 1.0  # 标准差

# 创建正态分布对象
normal_dist = Normal(mu, sigma)
```

用matplotlib绘制正态分布对象

```python
import matplotlib.pyplot as plt
# 生成一些数据点
x = torch.linspace(-5, 5, 10000)
# 得到概率密度值（对概率密度函数取对数，这样可以将乘法和除法转化为加法和减法，从而避免数值下溢。）
pdf_values = normal_dist.log_prob(x).exp()

# 绘制正态分布曲线
plt.plot(x, pdf_values, label='Normal Distribution (mu=0, sigma=1)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Normal Distribution')
plt.legend()
plt.grid(True)
plt.show()
```
